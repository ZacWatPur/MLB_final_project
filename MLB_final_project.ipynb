```python
# function to get and display divisional standings

def get_division_standings(league, division):
    standings = pd.DataFrame(statsapi.standings_data(league)[division]['teams'])
    standings = standings[['name', 'div_rank', 'w', 'l', 'gb', 'elim_num', 'wc_rank', 'wc_gb', 'wc_elim_num']]
    return standings

leagues = 103
for ecw in (200,201,202,203,204,205):
    if ecw == 203:
        leagues = 104
    div_standings = get_division_standings(leagues, ecw)
    display(div_standings)
```

```python
def season_stats(mlb_season):
    hr_leaders = statsapi.league_leaders('homeruns', season = mlb_season, limit = 10)
    hr_leaders = hr_leaders.split('\n')
    avg_leaders = statsapi.league_leaders('avg', season = mlb_season, limit = 10)
    avg_leaders = avg_leaders.split('\n')
    obp_leaders = statsapi.league_leaders('obp', season = mlb_season, limit = 10)
    obp_leaders = obp_leaders.split('\n')
    slg_leaders = statsapi.league_leaders('slg', season = mlb_season, limit = 10)
    slg_leaders = slg_leaders.split('\n')
    ops_leaders = statsapi.league_leaders('ops', season = mlb_season, limit = 10)
    ops_leaders = ops_leaders.split('\n')
    w_leaders = statsapi.league_leaders('wins', season = mlb_season, limit = 10)
    w_leaders = w_leaders.split('\n')
    l_leaders = statsapi.league_leaders('losses', season = mlb_season, limit = 10)
    l_leaders = l_leaders.split('\n')
    era_leaders = statsapi.league_leaders('era', season = mlb_season, limit = 10)
    era_leaders = era_leaders.split('\n')
    ip_leaders = statsapi.league_leaders('inningsPitched', season = mlb_season, limit = 10)
    ip_leaders = ip_leaders.split('\n')
    #k_leaders = statsapi.league_leaders('strikeOuts', season = mlb_season, limit = 10)
    #k_leaders = k_leaders.split('\n')
    #bb_leaders = statsapi.league_leaders('baseOnBalls', season = mlb_season, limit = 10)
    #bb_leaders = bb_leaders.split('\n')
    return hr_leaders, avg_leaders, obp_leaders, slg_leaders,ops_leaders, w_leaders, l_leaders, era_leaders, ip_leaders#, k_leaders, bb_leaders

season_stats(mlb_season)
```
